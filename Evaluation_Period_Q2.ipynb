```python
import pandas as pd
import math
AA=pd.DataFrame(columns=['criteria','weights','vendor1', 'vendor2','scoreA','scoreB'])
print(AA)
```

```python
AA['criteria']=['Cost','Scalability ','Security features','Customization option','Maintenance cost']
print(AA)
```

```python
AA['weights']=[20,8,15,7,12]
print(AA)
```

```python
AA['vendor1']=[5,6,7,9,8]
AA['vendor2']=[7,4,7,3,2]

print(AA)
```

```python
for index, row in AA.iterrows():
  AA['scoreA'][index]=row['vendor1']*row['weights']
  AA['scoreB'][index]=row['vendor2']*row['weights']
print(AA)
```

```python
totalScoreA=0
totalScoreB=0
for index, row in AA.iterrows():
  totalScoreA+=row['scoreA']
  totalScoreB+=row['scoreB']

print ('the total score for A 1s {:.2f} and for B is {:.2f}'.format (totalScoreA, totalScoreB))
```

```python
import numpy as np
import matplotlib.pyplot as plt
axisLocations = np.linspace(start=0, stop=2 * np.pi, num=len(AA['vendor1']), endpoint=False)
axisLocations = np.concatenate((axisLocations,[axisLocations[0]]))
ratingsA=AA['vendor1']
ratingsB=AA['vendor2']
ratingsACircular=np.concatenate((ratingsA,[ratingsA[0]]))
ratingsBCircular=np.concatenate((ratingsB,[ratingsB[0]]))
plt.figure(figsize=(8,8))
plt.subplot(polar=True)
plt.plot(axisLocations,ratingsACircular,'r-', linewidth=2, label='vendor1')
plt.plot(axisLocations,ratingsBCircular,'y-', linewidth=2, label='vendor2')
plt.title('Project options comparison', size=15)
lines, labels =plt.thetagrids(np.degrees(axisLocations), labels=AA['criteria'])
plt.legend()
plt.show()
```
