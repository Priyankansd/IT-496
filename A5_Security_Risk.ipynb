```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as plt

riskDf=pd.DataFrame(columns=['risk','likelihood','impact','score'])
riskDf['risk'] =['Unusual behavior from user accounts','Unauthorized insiders trying to access servers and data','Traffic sent to or from unknown locations', 'Anomalies in outbound network traffic']
riskDf['likelihood']=[0.8,0.7,0.2,0.5]
riskDf['impact']=[3,6,5,4]
riskDf['score']=riskDf['impact']*riskDf['likelihood']
scoresPivot=riskDf.pivot('impact','likelihood','score')
print(scoresPivot)
labelsPivot=riskDf.pivot('impact','likelihood','risk')
print(labelsPivot)

sns.set(rc={'figure.figsize':(15,8)})
p1=sns.heatmap(scoresPivot, cmap="RdYlGn",annot=labelsPivot,fmt='')
```
